```python
from sklearn.datasets import fetch_california_housing
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# Load the dataset
X, y = fetch_california_housing(return_X_y=True)

# Define the pipeline with preprocessing (StandardScaler) and modeling (KNeighborsRegressor)
pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Step 1: Scale the features
    ('model', KNeighborsRegressor())  # Step 2: Apply the model
])

# Fit the pipeline on the training data
pipeline.fit(X, y)

# Predict with the trained pipeline
predictions = pipeline.predict(X)

print(predictions)
```
